```python
data['date'] = pd.to_datetime(data['date'])
data
#converting date from object to datetime
```

```python
data.info()
```

```python
data['price'] = pd.to_numeric(data['price'], errors='coerce')
#converting price from object to float
```

```python
data.info()
```

```python
data['usdprice'] = pd.to_numeric(data['usdprice'], errors='coerce')
data['latitude'] = pd.to_numeric(data['latitude'], errors='coerce')
data['longitude'] = pd.to_numeric(data['longitude'], errors='coerce')
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
# Remove any rows with 0 or missing prices
data = data[data['price'] > 0].dropna(subset=['price'])
data.tail()


```

```python
#Group by commodity and get the average price
top_10 = data.groupby('commodity')['price'].mean().sort_values(ascending=False).head(10)

#Plot
plt.bar(top_10.index, top_10.values, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Most Expensive Commodities (Average)")
plt.ylabel("Price (NGN)")
plt.show()
```

```python
#Scatter Plot
maize_data = data[data['commodity'] == 'Maize']

#Plot
plt.scatter(maize_data['date'], maize_data['price'], alpha=0.5, s=10)
plt.title("Price of Maize Over Time")
plt.xlabel("Year")
plt.ylabel("Price (NGN)")
plt.show()
```

```python
plt.hist(data['price'], bins=30, color='seagreen', edgecolor='black')
plt.yscale('log')
plt.title("How Prices are Distributed")
plt.xlabel("Price Range")
plt.ylabel("Frequency (Log Scale)")
plt.show()
```

```python
data.boxplot(column='price', by='category', figsize=(10,6))
plt.yscale('log')
plt.xticks(rotation=45)
plt.title("Price Ranges by Category")
plt.suptitle("")
plt.show()
```

```python

data['year'] = data['date'].dt.year


pivot = data.pivot_table(index='commodity', columns='year', values='price', aggfunc='mean').head(10)

#Plot
plt.imshow(pivot, cmap='YlOrRd', aspect='auto')
plt.colorbar(label='Avg Price')

# Add labels so we know what is what
plt.yticks(range(len(pivot.index)), pivot.index)
plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=45)
plt.title("Price Heatmap (Top 10 Items)")
plt.show()
```
